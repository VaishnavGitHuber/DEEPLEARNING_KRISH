```python
# importing necessory libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# Loading the dataset
(X_train,y_train),(X_test,y_test) = tf.keras.datasets.mnist.load_data()
```

```python
# shape of the dataframe
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
# first image
plt.imshow(X_train[0])
plt.show()
print("Label: ",y_train[0])
```

```python
X_train[0]
```

```python
# applying the scaling
X_train = X_train / 255
X_test = X_test / 255
```

```python
# reshaping the dataframe / flattening
X_train = X_train.reshape((60000, 28*28))
X_test = X_test.reshape((10000, 28*28))
```

```python
# Building ANN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU, LeakyReLU, ELU
from tensorflow.keras.layers import Dropout
```

```python
# Model : Architechture
classifier = Sequential()

# input layer
classifier.add(Dense(units = 784, activation='relu'))

# hidden layer 1
classifier.add(Dense(units = 324, activation='relu'))
classifier.add(Dropout(0.3))

# hidden layer 2
classifier.add(Dense(units=72, activation='relu'))
classifier.add(Dropout(0.3))

# output layer
classifier.add(Dense(units=10, activation='softmax'))
```

```python
# compiling the model
classifier.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Early stopping
import keras

early_stopping = keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.005,
    patience=0,
    verbose=0,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
# model training
model_history = classifier.fit(X_train, y_train, epochs=10,
               batch_size=100, callbacks=early_stopping, validation_split=0.3)
```

```python
# Accuracy Visualisation
plt.title("Accuracy Visualisation")
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```

```python
# Accuracy Visualisation
plt.title("Accuracy Visualisation")
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.show()
```

```python
# predictions
predictions = classifier.predict(X_test)
```

```python
# y_prediction label
y_label = [np.argmax(prediction) for prediction in predictions]
```

```python
# Model Validation
from sklearn.metrics import accuracy_score, confusion_matrix

# accuracy score
accuracy_score(y_test, y_label)
```

```python
# confusion matrix
cf = confusion_matrix(y_test, y_label)

# heatmap visualisation
plt.title("Confusion Matrix")
sns.heatmap(cf, annot=True, cbar=False, fmt='0.0f')
plt.show()
```
